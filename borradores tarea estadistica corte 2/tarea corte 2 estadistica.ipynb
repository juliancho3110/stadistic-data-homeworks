Estadistica Descriptiva: es una de las ramas encargadas de mostrar resultados de multiples proyectos los datos de investigacion a traves de diversas herramientas tales como graficas, cuadros y tablas en la cual se puede optimizar los resultados mostrados para ser mas eficaces en la metodologia de investigacion de cualquier unidad investigativa 

Estadistica Inferencial: es una de las ramas de la estadistica que nos va a permitir hacer deducciones a partir de una muestra de datos determinada por una poblacion en especifico y asi desarrollar varias estrategias y conclusiones a partir de una investigacion en curso.
ademas es una de las ramas por las cuales se hace modelos de machine learning rama que ayuda a predecir modelos lineales y regresivos para  valga la redundancia predecir y proponer soluciones ante una determinada problematica 


Poblacion: es el conjunto de elementos que son objeto de estudio a la hora de analizar e interpretar algun proyecto

Muestra: es un subconjunto de elementos del cual proviene de la poblacion y son  datos muy especificos para hacer analisis que puedan generar conclusiones mas veraces por ejemplo si queremos analizar  tendencias y comportamientos de una poblacion de 100 habitantes la muestra puede ser de 20 personas de esas 100 que estaban siendo analizadas originalmente 

Variable Cuantitativa: es el tipo de dato que unicamente se puede medir numericamente de las cuales se puedan hacer operaciones aritmeticas por ejemplo los datos proveidos en nuestra tarea son una variable cuantitativa (20 datos) para analizar

Variable Cualitativa: es el tipo de datos que es determinado por caracteristicas o categorias especificas no numericas ya sean personas, objetos o situaciones 

## PARTE 2: Analisis exploratorio de los datos 

Calcule: 
- media
- mediana
- moda

Calcule las medidas de variabialidad:
- Rango
- varianza
- desviacion estandar

Organizar los datos en: 
- tablas de frecuencias
- grafico (barras o histograma)



````python
rendimiento_academico=[3.5,4.0,2.8,3.9,4.5,3.2,3.8,4.2,2.5,3.7,4.8,3.0,3.6,4.1,2.9,3.3,4.4,3.1,3.9,4.0]

datos=np.array(rendimiento_academico)
dato_repetido=pd.Series(datos)

media= np.mean(datos)
mediana= np.median(datos)
moda= dato_repetido.mode().to_list()
varianza= np.var(datos)
desviacion=np.std(datos)
rango= np.ptp(datos)
minimo=np.min(datos)
maximo=np.max(datos)



print("="*45)
print("reporte estadistico")
print("="*45)
print(f"media(promedio): {media}")

print(f"mediana: {mediana}")
print(f"moda: {moda}")
print(f"varianza: {varianza:.2f}")
print(f"desviacion estandar: {desviacion:.2f}")
print(f"rango: [{minimo}, {maximo}] {rango}")
print("="*45)


````

GRAFICANDO HISTOGRAMA

````python
plt.title("Distribucion de notas academicas")
plt.xlabel("Calificaciones")
plt.ylabel("cantidad de estudiantes(frecuencia)")
plt.hist(rendimiento_academico, bins=10, color='green', edgecolor='black', alpha=0.7)
plt.tight_layout()
plt.show()
````

ORGANIZANDO LOS DATOS EN TABLAS DE FRECUENCIAS


````python
# 1. Convertimos a Serie de Pandas
df_notas = pd.DataFrame(rendimiento_academico, columns=['Notas'])

# 2. Creamos los intervalos (bins)
# Usaremos 5 intervalos para que sea fácil de leer
intervalos = pd.cut(df_notas['Notas'], bins=5,precision=2)

# 3. Construimos la tabla de frecuencias
tabla_frec = df_notas.groupby(intervalos, observed=True).agg(
    frec_absoluta=('Notas', 'count')
)

# 4. Agregamos frecuencia acumulada y porcentual
tabla_frec['frec_acumulada'] = tabla_frec['frec_absoluta'].cumsum()
tabla_frec['frec_relativa (%)'] = (tabla_frec['frec_absoluta'] / len(df_notas)) * 100

print(tabla_frec)
````
